# Chord Length

```python
import matplotlib.pyplot as plt
import numpy as np
import draw
from scipy import ndimage
import seaborn as sns
from tqdm import tqdm

%load_ext autoreload
%autoreload 2
```

```python


def get_BINARY(fname, box_L, diameter):
    gelx, gely, gelz = np.loadtxt(fname, skiprows = 2, usecols = (1,2,3), unpack = True)
    N = len(gelx)
    radius = diameter / 2
    
    new_g = np.zeros((3,N))
    new_g[0] = gelx
    new_g[1] = gely
    new_g[2] = gelz
    
    new_g += box_L
    
    new_g = diameter*new_g
    width = int(diameter * box_L * 2)
    box = np.zeros((width, width, width), dtype=np.uint8)
    binary = draw.simulate_binary_3d(box, new_g, radius)
    
    return binary

def get_PBC_L(slice_1d):
    pbc_L = 0
    in_pore = True
    while in_pore:
        if slice_1d[pbc_L] == 0:
            pbc_L +=1
        else:
            in_pore = False 
    return pbc_L

def get_PORE_START_IDs(slice_1d):
    startIDs = []
    for i, cell in enumerate(slice_1d[:-1]):
        if cell == 1 and slice_1d[i+1] == 0:
            startIDs.append(i+1)
    return startIDs

def get_CHORDS_1D(slice_1d):
    pbc_L = 0
    if slice_1d[0] == 0:
        pbc_L = get_PBC_L(slice_1d)
    
    startIDs = get_PORE_START_IDs(slice_1d)
    
    for chordStart in startIDs:
        chord_L = 0
        in_pore = True
        while in_pore:
            if chordStart + chord_L < len(slice_1d):
                if slice_1d[chordStart + chord_L] == 0:
                    chord_L +=1
                else:
                    in_pore = False
            else:
                chord_L += pbc_L
                in_pore = False
        chords.append(chord_L)
   
```

```python
diameter = 20
radius = diameter / 2
p = np.array([[radius,radius,radius],
             [radius,radius,radius],
             [radius,radius,radius],],dtype=float)

width = diameter + 1
box = np.zeros((width, width, width), dtype=np.uint8)
structure = (draw.simulate_binary_3d(box, p, radius))/3
plt.imshow(structure[int(radius)])
```

```python
import numpy as np
import draw
from scipy import ndimage
from tqdm import tqdm


def get_BINARY(fname, box_L, diameter):
    
    full_coords = np.loadtxt(fname, skiprows=10, usecols=(1,2,3,4))
    num_pins =0
    p_IDs = full_coords[:,0]
    indexID = []
    for i in range(len(p_IDs)):
        if p_IDs[i] == 1:
            indexID.append(i)
    num_pins = len(indexID)
    pin_coords = np.zeros((num_pins,3))
    for i, pinID in enumerate(indexID):
        pin_coords[i,:] = full_coords[pinID,1:]
    
    
    gelx, gely, gelz = pin_coords[:,0], pin_coords[:,1], pin_coords[:,2]
    N = len(gelx)
    radius = diameter / 2
    
    new_g = np.zeros((3,N))
    new_g[0] = gelx
    new_g[1] = gely
    new_g[2] = gelz
    
    new_g += box_L
    
    new_g = diameter*new_g
    width = int(diameter * box_L * 2)
    box = np.zeros((width, width, width), dtype=np.uint8)
    binary = draw.simulate_binary_3d(box, new_g, radius)
    
    return binary

def get_PBC_L(slice_1d):
    pbc_L = 0
    in_pore = True
    while in_pore:
        if slice_1d[pbc_L] == 0:
            pbc_L +=1
        else:
            in_pore = False 
    return pbc_L

def get_PORE_START_IDs(slice_1d):
    startIDs = []
    for i, cell in enumerate(slice_1d[:-1]):
        if cell == 1 and slice_1d[i+1] == 0:
            startIDs.append(i+1)
    return startIDs

def get_CHORDS_1D(slice_1d):
    pbc_L = 0
    if slice_1d[0] == 0:
        pbc_L = get_PBC_L(slice_1d)
    
    startIDs = get_PORE_START_IDs(slice_1d)
    
    for chordStart in startIDs:
        chord_L = 0
        in_pore = True
        while in_pore:
            if chordStart + chord_L < len(slice_1d):
                if slice_1d[chordStart + chord_L] == 0:
                    chord_L +=1
                else:
                    in_pore = False
            else:
                chord_L += pbc_L
                in_pore = False
        chords.append(chord_L)
    return
   



diameter = 20
radius = diameter / 2
p = np.array([[radius,radius,radius],
             [radius,radius,radius],
             [radius,radius,radius],],dtype=float)
width = diameter + 1
box = np.zeros((width, width, width), dtype=np.uint8)
structure = (draw.simulate_binary_3d(box, p, radius))/3

box_L = 27.5017
width = int(2*box_L * diameter)

for s in np.arange(1,11):
    print(f'running seed {s} ...')
    fname = f'initial_config_RAND_144000_1_{s}.dat '
    binary = get_BINARY(fname, box_L, diameter)
    print('dilating...')
    dil = ndimage.binary_dilation(binary, structure=structure).astype(int)
    print('eroding...')
    ero = ndimage.binary_erosion(dil, structure=structure).astype(int)
    
    
    print('measuring chords...')
    chords = []

    for j in tqdm(range(len(ero[:]))):
        for i in range(width):
            # X
            slice_1D = ero[j][:,i]
            if sum(slice_1D) != 0:
                get_CHORDS_1D(slice_1D)
            
            # Y
            slice_1D = ero[j][i,:]
            if sum(slice_1D) != 0:
                get_CHORDS_1D(slice_1D)

            # Z
            slice_1D = ero[:][i,j]
            if sum(slice_1D) != 0:
                get_CHORDS_1D(slice_1D)
    np.savetxt(f'rand_chords_{s}.txt', np.array(chords))
```
